## Reading in files

```python
df = pd.read_csv('events_post_cuts_1207.csv', sep = ',')
df
```

```python
zcmb_list = df['zCMB']
RA_list = df['RA']
DEC_list = df['DEC']
CID_list = df['CID']
```

## Calculating Peculiar Velocities

```python
pv_list =[]

for i in range(0,len(zcmb_list)):
    zcmb = zcmb_list[i]
    RA = RA_list[i]
    DEC = DEC_list[i]
    if zcmb >0.067:
        pv_list.append(0)
    else:
        pv_value = pv.calculate_pv(RA, DEC, zcmb, extrapolation=True)
        pv_list.append(pv_value)
```

```python
len(pv_list)
```

```python
header = ['CID','V_pec']

with open("peculiar_velocities.csv", 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter = ',')
    csvwriter.writerow(header)
    for i in range(0,len(pv_list)):
        row = [str(CID_list[i]),str(pv_list[i])]
        csvwriter.writerow(row)
```
