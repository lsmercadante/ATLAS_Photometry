```python
df = pd.read_csv('/Users/lilahskye/Desktop/Cosmology_Research/ATLAS_Project/events_roundedz_001.csv', sep = ',')
CID_list = df['CID']
print(len(CID_list))
```

```python
RA_best_list = []
DEC_best_list = []
z_best_list = []
sep_best_list = []
name_best_list = []
CID_pass_list = []


for CID in CID_list:
    f = open('NED/{SN_NAME}.txt'.format(SN_NAME = CID), 'r')
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline()
    f.readline() 

    name_list = []
    RA_list = []
    DEC_list = []
    z_list = []
    sep_list = []
    
    for line in f:
        line = line.strip()
        columns = line.split(sep = '|')
        name = columns[1]
        RA = columns[2]
        DEC = columns[3]
        type = columns[4]
        z = columns[6]
        sep = columns[9]
        if type =='G':
            name_list.append(name)
            RA_list.append(RA)
            DEC_list.append(DEC)
            z_list.append(z)
            sep_list.append(float(sep.replace(' ','')))

    #print(sep_list)
    if len(sep_list) == 0:
        print('List is Empty')
        continue

    else:
        min_sep = np.min(sep_list)
        min_index = sep_list.index(min_sep)
        print(min_index)
        name_best_list.append(name_list[min_index])
        RA_best_list.append(RA_list[min_index])
        DEC_best_list.append(DEC_list[min_index])
        z_best_list.append(z_list[min_index])
        sep_best_list.append(sep_list[min_index])
        CID_pass_list.append(CID)
        #name_best_list.append(name_list[min_index])
        print(name_list[min_index])
```
