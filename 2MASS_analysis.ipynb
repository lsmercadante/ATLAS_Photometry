```python
RA_pass_list = []
DEC_pass_list = []
z_pass_list = []
obj_name_pass_list = []
CID_pass_list = []

for i in range(0,len(aCID_list)):
    ATLAS_RA = aRA_list[i]
    ATLAS_DEC = aDEC_list[i]
    SN_name = aCID_list[i]
    RA_sep_list = []
    DEC_sep_list = []
    
    for RA in RA_list:
        RA_sep = abs(ATLAS_RA - float(RA))
        RA_sep_list.append(RA_sep)
    for DEC in DEC_list:
        DEC_sep = abs(ATLAS_DEC - float(DEC))
        DEC_sep_list.append(DEC_sep)
    RA_sep_min = np.min(RA_sep_list)
    DEC_sep_min = np.min(DEC_sep_list)

    min_ra_index = RA_sep_list.index(RA_sep_min)
    min_dec_index = DEC_sep_list.index(DEC_sep_min)

    if min_ra_index != min_dec_index:
        print('Position Match NOT FOUND')
        continue
        
    else:
        if RA_sep_min < 0.0045 and DEC_sep_min < 0.003:
            print('Position Match FOUND')
        
            RA_min = RA_list[min_ra_index]
            DEC_min = DEC_list[min_ra_index]
            z_min = z_list[min_ra_index]
            obj_name_min = obj_name_list[min_ra_index]
            RA_pass_list.append(RA_min)
            DEC_pass_list.append(DEC_min)
            z_pass_list.append(z_min)
            obj_name_pass_list.append(obj_name_min)
            CID_pass_list.append(aCID_list[i])
        
        else:
            print('NO Match Found')
            continue
```
