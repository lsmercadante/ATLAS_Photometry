```python
import numpy as np
import pandas as pd
import math
import glob
```

```python
NED_result = pd.read_csv('NED_results.csv', sep = ',')
NED_result
```

```python
NED_pass = []
NED_z_pass = []
for i in range(0,len(NED_result['CID'])):
    CID = NED_result['CID'][i]
    z = NED_result['z'][i]
    if math.isnan(z):
        print('No z value')
    else:
        NED_pass.append(CID)
        NED_z_pass.append(z)
```

```python
print(len(NED_pass))
```

```python
SDSS_results = glob.glob('SDSS/*.csv')
SDSS_events = []
SDSS_z = []

for filename in SDSS_results:
    SN_name = filename.replace('.csv','').replace('SDSS/','')
    df = pd.read_csv(filename, sep = ',')
    z = df['z'][0]
    SDSS_events.append(SN_name)
    SDSS_z.append(z)
```

```python
SDSS_NED_overlap = []
SDSS_unique = []
for SN_name in SDSS_events:
    if SN_name in NED_pass:
        print(SN_name)
        SDSS_NED_overlap.append(SN_name)
        index_SDSS = SDSS_events.index(SN_name)
        index_NED = NED_pass.index(SN_name)
        print(SDSS_z[index_SDSS], NED_z_pass[index_NED])
    else:
        SDSS_unique.append(SN_name)
        
```

```python
len(SDSS_NED_overlap)
```

```python
print(len(SDSS_events) - len(SDSS_NED_overlap))
```

```python
DESI_results = glob.glob('DESI/*.csv')
DESI_events = []
DESI_z = []

for filename in DESI_results:
    SN_name = filename.replace('.csv','').replace('DESI/','')
    df = pd.read_csv(filename, sep = ',')
    z = df['z'][0]
    spectype = df['spectype'][0]
    print(spectype)
    if spectype=='GALAXY':
        DESI_events.append(SN_name)
        DESI_z.append(z)
```

```python
DESI_NED_overlap = []
for SN_name in DESI_events:
    if SN_name in NED_pass:
        print(SN_name)
        DESI_NED_overlap.append(SN_name)
        index_DESI = DESI_events.index(SN_name)
        index_NED = NED_pass.index(SN_name)
        print(DESI_z[index_DESI], NED_z_pass[index_NED])
```
